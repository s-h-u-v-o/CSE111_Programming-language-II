```python
def calc_bmi(h, w) :
    bmi = round(w/(h/100)**2, 1)
    if bmi < 18.5 :
        print(f'Score is {bmi}. You are Underweight')
    elif 18.5 <= bmi <= 24.9 :
        print(f'Score is {bmi}. You are Normal')
    elif 25 <= bmi <= 30 :
        print(f'Score is {bmi}. You are Overweight')
    elif 30 < bmi :
        print(f'Score is {bmi}. You are Obese')

calc_bmi(175, 96)
```

```python
def palindrome(x) :
    y = ''
    for i in x :
        if 65 <= ord(i) <= 90 or 97 <= ord(i) <= 122 :
            y += i
        else :
            continue
    if y == y[-1::-1] :
        print('Palindrome')
    else :
        print('Not a palindrome')

palindrome('madam')
palindrome('hello')
palindrome('nurses run')
```

```python
def grammarly(text) :
    new = ''
    if 97 <= ord(text[0]) <= 122 :
        new += chr(ord(text[0]) - 32)
    else :
        new += text[0]
    i = 1
    while i < len(text) :
        if (text[i] == '.' or text[i] == '?' or text[i] == '!') and i != len(text)-1 :
            if 97 <= ord(text[i+2]) <= 122 :
                new += text[i] + ' ' + chr(ord(text[i+2]) - 32)
            else :
                new += text[i] + ' ' + text[i+2]
            i += 2
        elif text[i] == 'i' and text[i-1] == ' ' and text[i-1] == ' ' :
            new += 'I'
        else :
            new += text[i]
        i += 1
    return new

grammarly('hey! how are you? i am fine.')
```

```python
def hospital_fee(**d1) :
    d2, x = {}, -99999
    for i, j in d1.items() :
        if j in d2 :
            d2[j] += ', ' + i
        else :
            d2[j] = i
    for i in d2.keys() :
        if i >= x :
            x = i
    return x, d2[x]

max_amount, max_payer = hospital_fee(Neymar = 1000, Dembele = 600, Reus = 500, Bale = 1000)
print(f'Highest fee was {max_amount} tk which was paid by {max_payer}.')
```

```python
def is_james_bond(l1) :
    l2 = []
    for i in l1 :
        if i == 0 or i == 7 :
            l2.append(i)
    if l2 == [0, 0, 7] :
        return True
    else :
        return False

is_james_bond( [1, 0, 2, 0, 4, 7, 5] )
```

```python
def assign_students_to_sections(sec, *names) :
    dict = {}
    for i in sec :
        dict[i] = []
    for i in range(len(names)) :
        sum = 0
        for j in names[i] :
            sum += ord(j)
        mod = sum%5
        dict[sec[mod]] += [names[i]]
    return dict

assign_students_to_sections ('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace')
```

```python
def username_generator(first_name, last_name, student_id, middle_name='') :
    username = first_name[:3].upper() + middle_name + last_name[-3:] + '_' + str(student_id)[-4:]
    return username

first_name, middle_name, last_name, student_id = input ("First Name: "), input("Middle Name: "), \
                                                input ("Last Name: "), int (input ("Student ID: "))
print(username_generator (first_name, last_name, student_id, middle_name))
```

```python
def key_generator(*lst) :
    keys = []
    for name in lst :
        key = ''
        key += name[0].lower()
        for i in range(len(name)-2, 0, -1) :
            key += str(ord(name[i]))
        key += name[-1].upper()
        keys.append(key)
    return keys
key_list = key_generator("Alex", "Bob", "Trudy")
print(key_list)
```

```python
import random

def playRockPaperScissor(rounds) :
    pl_score, pc_score = 0, 0
    pairs = {'rock': 'scissor', 'paper': 'rock', 'scissor': 'paper'}
    for i in range(rounds) :
        player = input('Player: ').lower()
        computer = random.choice(['rock', 'paper', 'scissor'])
        print(f"Computer: {computer}")
        if player == computer :
            continue
        elif pairs[player] == computer :
            pl_score += 1
        else :
            pc_score += 1
    if pl_score > pc_score :
        winner = "You have won the game!"
    elif pl_score < pc_score :
        winner = "Computer has won the game!"
    else :
        winner = "It's a tie!"
    return pl_score, pc_score, winner

rounds = int(input("rounds: "))
pl_score, pc_score, winner = playRockPaperScissor(rounds)
print(f"Your Score: {pl_score}")
print(f"Computer's Score: {pc_score}")
print(winner)
```
