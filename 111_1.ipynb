```python
lst = []
while True :
    x = input()
    if x == 'STOP' :
        break
    lst.append(int(x))
for i in range(len(lst)) :
    count = 1
    if lst[i] == None :
        continue
    for j in range(i+1, len(lst)) :
        if lst[i] == lst[j] :
            count += 1
            lst[j] = None
    print(f'{lst[i]} - {count} times')
```

```python
n, lst = int(input()), []
max_sum, max_list = 0, []
for i in range(n) :
    sum = 0
    lst = [int(i) for i in input().split(' ')]
    for j in range(len(lst)) :
        sum += lst[j]
    if sum > max_sum :
        max_sum = sum
        max_list = lst
print(f'{max_sum}\n{max_list}')
```

```python
while True :
    y = input()
    if y == 'STOP' :
        break
    lst = [int(i) for i in y.split(' ')]
    for i in range(len(lst)-1) :
        d = abs(lst[i] - lst[i+1])
    if 1<= d <= len(lst)-1 :
        print('UB Jumper')
    else :
        print('Not UB Jumper')
```

```python
def sorting(x) :
    for i in range(len(x)-1) :
        for j in range(i+1, len(x)) :
            if x[i] > x[j] :
                x[i], x[j] = x[j], x[i]
    return x
x = 'Bracu1234'
lower, upper, even, odd = [], [], '', ''
for i in range(len(x)) :
    if 97 <= ord(x[i]) <= 122 :
        lower.append(x[i])
    elif 65 <= ord(x[i]) <= 90 :
        upper.append(x[i])
    elif 48 <= ord(x[i]) <= 57 :
        if int(x[i])%2 == 0 :
            even += x[i]
        else :
            odd += x[i]
lower, upper = sorting(lower), sorting(upper)
chr = ''
for i in lower :
    chr += i
for i in upper :
    chr += i
print(chr+odd+even)
```

```python
d1 = {'a': 100, 'b': 100, 'c': 200, 'd': 300}
d2 = {'a': 300, 'b': 200, 'd': 400, 'e': 200}
l1, l2 = [], []
for i, j in d1.items() :
    for p, q in d2.items() :
        if i == p :
            d1.update({i: j+q})
for p, q in d2.items() :
    if p not in d1.keys() :
        d1.update({p: q})
for val in d1.values() :
    l1.append(val)
l1.sort()
for i in l1 :
    if i in l1 and i not in l2 :
        l2.append(i)
print(d1)
print(tuple(l2))
```

```python
def generate_key_presses(message):
    key_mapping = {'1': '.,?!:','2': 'ABC','3': 'DEF','4': 'GHI','5': 'JKL',\
    '6': 'MNO','7': 'PQRS','8': 'TUV','9': 'WXYZ','0': ' '}

    result = ''
    for char in message.upper():
        for key, value in key_mapping.items():
            if char in value:
                result += str(value.index(char) + 1) * (value.index(char) + 1)

    return result

user_input = "Hello"

output = generate_key_presses(user_input)
print(output)
```
