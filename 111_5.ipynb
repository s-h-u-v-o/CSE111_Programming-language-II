```python
class Student :
    def __init__(self, name, id, dept = 'CSE') :
        self.name = name
        self.id = id
        self.dept = dept
    def dailyEffort(self, hour) :
        self.effort = hour
        if hour <= 2 :
            self.suggestion = 'Suggestion: Should give more effort!'
        elif hour <= 4 :
            self.suggestion = 'Suggestion: Keep up the good work!'
        else :
            self.suggestion = 'Suggestion: Excellent! Now motivate others.'
    def printDetails(self) :
        print('Name:', self.name)
        print('ID:', self.id)
        print('Department:', self.dept)
        print(f'Daily Effort: {self.effort}')
        print(self.suggestion)

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
class Farmer :
    def __init__(self, a = None) :
        if a == None :
            print("Welcome to your farm!")
        elif type(a) == str :
            print('Welcome to your farm', a)
        elif type(a) == int :
            print('Welcome to your farm. Your farm ID is', a)
        self.crops, self.fishes = [], []
    def addCrops(self, *crops) :
        if len(crops) == 0 :
            print('No crop(s) added.')
        else :
            print(f'{len(crops)} crop(s) added.')
            for i in crops :
                self.crops.append(i)
    def addFishes(self, *fishes) :
        if len(fishes) == 0 :
            print('No fish(s) added.')
        else :
            print(f'{len(fishes)} fish(s) added.')
            for i in fishes :
                self.fishes.append(i)
    def showGoods(self) :
        if len(self.crops) == 0 :
            print('You don\'t have any crops')
        else :
            print(f'You have {len(self.crops)} crops:')
            crops = ''
            for i in self.crops :
                crops += i + ', '
            print(crops[:-2:])
        if len(self.fishes) == 0 :
            print('You don\'t have any fishes')
        else :
            print(f'You have {len(self.fishes)} fishes:')
            fishes = ''
            for i in self.fishes :
                fishes += i + ', '
            print(fishes[:-2:])

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
class TaxiLagbe :
    def __init__(self, id, area) :
        self.id = id
        self.area = area
        self.lst = []
    def addPassenger(self, *name_fare) :
        for i in name_fare :
            if len(self.lst) < 4 :
                self.lst.append(i)
                x, y = i.split('_')
                print(f'Dear {x}! Welcome to TaxiLagbe.')
            else :
                print('Taxi Full! No more passengers can be added.')
    def printDetails(self) :
        print('Trip info for Taxi number:', self.id)
        print(f'This taxi can only cover the {self.area} area.')
        print('Total passengers:', len(self.lst))
        name, fare = '', 0
        for i in self.lst :
            x, y = i.split('_')
            name += x + ', '
            fare += int(y)
        print(f'Passenger lists:\n{str(name)[:-2]}')
        print('Total collected fare:', fare, 'Taka')

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
class Sphere :
    def __init__(self, id, r = 1, c = 'White') :
        self.id = id
        self.r = r
        self.c = c
        self.v = (4/3) * 3.1416 * r**3
    def merge_sphere(self, *args) :
        print('Spheres are being merged')
        for i in args :
            self.v += i.v
            if self.c != i.c :
                self.c = 'Mixed Color'
    def printDetails(self) :
        print('Sphere ID:', self.id)
        print('Color:', self.c)
        print('Volume:', self.v)

sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
class Foodie :
    def __init__(self, name) :
        self.menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}
        self.name = name
        self.lst = []
    def order(self, *args) :
        orders, string = [], ''
        for i in args :
            x, y = i.split('-')
            self.lst.append(x)
            orders.append(x)
            orders.append(y)
            if x in self.menu :
                z = self.menu[x]
                orders.append(z)
                orders.append(z*int(y))
        for i in range(0, len(orders), 4) :
            o = orders[i]
            q = str(orders[i+1])
            p = str(orders[i+2])
            t = str(orders[i+3])
            temp = 'Ordered - '+o+', quantity - '+q+', price (per Unit) - '+p+'.\n'+'Total price - '+t+'\n'
            string += temp
        print(string)
    def show_orders(self) :
        print('Items:', self.lst)
    def pay_tips(self, tip = 0) :
        if tip == 0 :
            print('No tips to the waiter.')
        else :
            print(f'Gives {tip}/- tips to the waiter.')


f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
class Department :
    def __init__(self, dept='ChE Department', sec=5) :
        self.dept = dept
        self.sec = sec
        print(f'The {self.dept} has {self.sec} sections.')
    def add_students(self, *args) :
        self.avg = 0
        self.count = 0
        count = 0
        if len(args) == self.sec :
            for i in args :
                count += i
            count = round(count/len(args), 2)
            self.avg += count
            self.count = (self.avg*self.sec)
            print(f'The {self.dept} has an average of {count} students in each section.')
        else :
            print(f'The {self.dept} doesn\'t have {len(args)} sections.')
    def merge_Department(self, *depts) :
        for i in depts :
            print(f'{i.dept} is merged to {self.dept}.')
            self.count += (i.avg*i.sec)
        return(f'Now the {self.dept} has an average of {round(self.count/self.sec, 1)} students in each section.')

d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify :
    def __init__(self, name = 'Guest') :
        self.name = name
        if self.name == 'Guest' :
            print('Welcome to Shopidify')
        else :
            print(f'Welcome {self.name} to Shopidify')
    def add_to_cart(self, *orders) :
        self.order = []
        self.quantity = []
        for i in orders :
            if i in self.order :
                self.quantity[self.order.index(i)] += 1
            else :
                if type(i) == list :
                    for j in range(0, len(i), 2) :
                        self.order.append(i[j])
                        self.quantity.append(i[j+1])
                else :
                    if len(orders) == 1 :
                        self.order.append(i)
                        self.quantity.append(1)
                    else :
                        if type(i) == str :
                            self.order.append(i)
                        else :
                            self.quantity.append(i)
    def display_cart(self) :
        print(f'Items in the cart for {self.name}:')
        for i in range(len(self.order)) :
            print(f'- {self.order[i]}: {self.quantity[i]}x')
    def checkout(self) :
        print('Checkout completed for', self.name)
    def display_history(self) :
        print(f'Purchase history for {self.name}:\nTransaction 1:')
        for i in range(len(self.order)) :
            print(f'- {self.order[i]}: {self.quantity[i]}x')

guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
class Author :
    def __init__(self, author = '') :
        self.author = author
        self.count = 0
        self.dict = {}
    def setName(self, author) :
        self.author = author
    def addBook(self, name, genre) :
        if self.author == '' :
            print('A book can not be added without author name')
        else :
            if genre in self.dict.keys() :
                self.dict[genre] += ', ' + name
            else :
                self.dict[genre] = name
            self.count += 1
    def printDetail(self) :
        print('Number of Book(s) :', self.count)
        print('Author Name:', self.author)
        for i, j in self.dict.items() :
            print(f'{i}: {j}')

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
