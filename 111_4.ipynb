```python
class Customer :
    def __init__(self) :
        print('Welcome to ABC Memorial Park')
        self.count = 0
        self.price = 0
    def buyTicket(self, name, age) :
        if self.count < 3 :
            if age > 10 :
                self.price += 100
            else :
                self.price += 50
            self.count += 1
            print(f'Successfully purchased a ticket for {name}!')
        else :
            print('You can\'t buy more than 3 tickets')
    def showDetails(self) :
        print('Amount of tickets:', self.count)
        print(f'Total price: {self.price} Taka')

print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
class Panda :
    def __init__(self, name, gender, age) :
        self.name = name
        self.gender = gender
        self.age = age
    def sleep(self, time) :
        self.time = time
        if 2 < self.time < 6 :
            return(f'{self.name} sleeps {self.time} hours \
daily and should have Mixed Veggies')
        elif 5 < self.time < 9 :
            return(f'{self.name} sleeps {self.time} hours \
daily and should have Eggplant & Tofu')
        elif 8 < self.time < 12 :
            return(f'{self.name} sleeps {self.time} hours \
daily and should have Broccoli Chicken')
        else :
            return(f'{self.name}\'s duration is unknown thus \
should have only bamboo leaves')

panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female",3)
panda3 = Panda("Ming Ming", "Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep(13))
```

```python
class StudentDatabase :
    def __init__(self, name, id) :
        self.name = name
        self.id = id
        self.grades = {}
    def calculateGPA(self, lst, session) :
        courses, gpa, d1, d2 = [], 0, {}, {}
        for i in lst :
            courses.append(i[:-5:])
            gpa += float(i[-3::])/len(lst)
        d2[tuple(courses)] = gpa
        d1[session] = d2
        self.grades.update(d1)
    def printDetails(self) :
        print('Name:', self.name)
        print('ID:', self.id)
        for i, j in self.grades.items() :
            print(f'Courses taken in {i}:')
            for k, l in j.items() :
                for m in k :
                    print(m)
                print('GPA:', l)
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
class abcTech :
    def __init__(self, name, desig, dept) :
        print('Welcome to abcTech, Tamim Hasan!')
        self.name = name
        self.desig = desig
        self.dept = dept
        self.skills = ''
        self.fr_works = ''
        self.salary = 0
    def addProgrammingSkills(self, l1) :
        for i in l1 :
            self.skills += i+', '
        self.skills = self.skills
    def addFrameworks(self, l2) :
        for i in l2 :
            self.fr_works += i+', '
        self.fr_works = self.fr_works
    def printInfo(self) :
        print(f'''Name: {self.name}
Designation:  {self.desig}
Department: {self.dept}
Programming Skills:  {self.skills[:-2:]}
Frameworks:  {self.fr_works[:-2:]}''')
    def calculateSalary(self, base, wh) :
        if wh > 144 :
            self.salary = base + 800*(wh-144)
        else :
            self.salary = base
        return self.salary
print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
class Store :
    def __init__(self, balance) :
        self.balance = balance
        self.total_items = 0
        self.items = ''
        self.details = {}
        print('New branch created!')
    def viewAllItems(self) :
        if self.items != '' :
            print(f'All Items: {self.items[:-2:]}')
        else :
            print('There are no items in your inventory')
    def viewAllItemDetails(self) :
        print(self.details)
    def add_item(self, lst) :
        print(f'Item added: {lst[0]}')
        self.balance -= lst[1]*lst[2]
        self.items += lst[0] + ', '
        self.stock = lst[1]
        self.buying_price = lst[2]
        self.selling_price = lst[3]
        self.details[lst[0]] = {'stock': lst[1], 'buying_price': lst[2], 'selling_price': lst[3]}
    def sell_item(self, item, quantity) :
        if quantity > self.details[item]['stock'] :
            print(f'Sorry! ChaCha Noodles is not available at your desired quantity. Currently we have: {self.details[item]["stock"]}')
        else :
            self.balance += self.details[item]['selling_price']*quantity
    def restock_item(self, item, quantity) :
        self.details[item]['stock'] += quantity
        self.balance -= self.details[item]['buying_price']*quantity

print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```

```python
class SentenceAnalyzer :
    def __init__(self, sent = '') :
        self.sent = sent.split(' ')
    def set_sentence(self, sent) :
        self.sent = sent.split(' ')
    def get_word_count(self, num = 0) :
        if num == 0 :
            print('Number of words in the sentence:', len(self.sent))
        else :
            l_count = 0
            for i in self.sent :
                if len(i) == num :
                    l_count += 1
            print(f'Number of words with {num} characters in the sentence: {l_count}')

analyzer1 = SentenceAnalyzer()

analyzer1.set_sentence("That's an easy one")
print("1--------------------------")
analyzer1.get_word_count()
print("2--------------------------")
analyzer2 = SentenceAnalyzer("Like I said it's easy")

print("3--------------------------")
analyzer2.get_word_count()

print("4--------------------------")
analyzer2.get_word_count(4)
print("5--------------------------")
analyzer1.get_word_count(5)
```

```python
class Author :
    def __init__(self, author = '') :
        self.author = author
        self.count = 0
        self.dict = {}
    def setName(self, author) :
        self.author = author
    def addBook(self, name, genre) :
        if self.author == '' :
            print('A book can not be added without author name')
        else :
            if genre in self.dict.keys() :
                self.dict[genre] += ', ' + name
            else :
                self.dict[genre] = name
            self.count += 1
    def printDetail(self) :
        print('Number of Book(s) :', self.count)
        print('Author Name:', self.author)
        for i, j in self.dict.items() :
            print(f'{i}: {j}')

a1 = Author()
print("===============================")
a1.addBook("Ice", "Science Fiction")
print("===============================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("===============================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon","Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("===============================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("===============================")
a2.printDetail()
print("===============================")
```
