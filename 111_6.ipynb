```python
class Teacher :
    def __init__(self, name, dept) :
        self.name = name
        self.dept = dept
        self.Course = []
    def addCourse(self, c) :
        self.Course.append(c)
    def printDetail(self) :
        print(f'====================================\nName: {self.name}\nDepartment: {self.dept}\nList of courses\n====================================')
        for i in self.Course :
            print(i.name)
        print('====================================')
class Course :
    def __init__(self, name) :
        self.name = name

t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
class Student :
    def __init__(self, name, id, cg) :
        self.lst = []
        self.name = name
        if id in self.lst :
            print('Student with the same ID already exists, Please try with another ID')
        else :
            self.id = id
        self.cg = cg
        self.lst += [self.name, self.id, self.cg]
    def setId(self, id) :
        self.lst[-2] = id

class Department :
    def __init__(self, dept) :
        self.dept = dept
        self.students = []
    def findStudent(self, id) :
        if id not in self.students :
            print('Student with this ID doesn\'t exist, Please give a valid ID')
        else :
            i = self.students.index(id)
            print('Student info:')
            print('Student Name:', self.students[i-1])
            print('ID:', self.students[i])
            print('CGPA:', self.students[i+1])
    def addStudent(self, *args) :
        for i in args :
            if i.lst[1] in self.students :
                print('Student with the same ID already exists, Please try with another ID')
            else :
                for j in i.lst :
                    self.students.append(j)
                print('Welcome to CSE department,', i.lst[0])
    def details(self) :
        print('Department Name: CSE')
        print('Number of students:', int(len(self.students)/3))
        print('Details of the students:')
        for i in range(0, len(self.students), 3) :
            print(f'Student name: {self.students[i]}, ID: {self.students[i+1]}, cgpa: {self.students[i+2]}')


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```
